## Task 3: Load the Data

```python
random.seed(42)
num_rows = 2500  # Generates 2,500 rows

# 1. Define "Messy" Data Pools
job_titles = [
    "Data Analyst", "DATA ANALYST", "Sr. Data Analyst", "Senior Data Analyst", "Data Analyst II",
    "Jr. Data Analyst", "Junior Data Analyst", "Data Analytics Specialist", "BI Analyst",
    "Data Analyst - Remote", "Product Data Analyst", "Marketing Data Analyst", "Data Analyst!!!"
]

salaries = [
    "$60,000 - $80,000", "$90,000", "110000", "120k", "85k - 105k", "70000 USD", "€65,000", "%50,000", "65000"
    "Competitive", "DOE", "Negotiable", "$50/hr", "140,000.00", "80-90k", np.nan, "95000"
]

locations = ["New York, NY", "San Francisco, CA", "Dallas, TX", "Remote", "US", "London, UK", "Austin, TX", "Anywhere", np.nan,"Houston,TX", "Huntsville,AL"]
remote_options = ["Yes", "No", "Hybrid", "True", "False", "1", "0", "Partial", "Temporarily Remote", ]
sizes = ["10000+", "501-1000", "1-10", "Large", "Medium", "Startup", "Unknown", "500"]
experience = ["Entry-level", "Senior", "Junior", "Principal", "0-2 years", "5+ years", "Sr.", "Jr", "SE", "MI", "EN", "0-5 yrs"]
employment = ["Full-time", "Part-time", "Contract", "Temporary", "FT", "PT", "C2C", "W2", "Freelance"]
skills_pool = ["SQL", "Python", "Excel", "Tableau", "Power BI", "R", "AWS", "Google Cloud", "Azure", ""]

# 2. Helper Function for Messy Skills
def get_messy_skills():
    if random.random() < 0.05: return np.nan
    selected = random.sample(skills_pool, k=random.randint(2, 5))
    separator = random.choice([", ", ",", ";", " | ", "/", " and "])
    # Randomly return as a string list representation or just text
    return str(selected) if random.random() < 0.5 else separator.join(selected)

# 2.5 Generate 1–3 locations per row
def get_random_locations(max_locations=3):
    # Filter out NaN properly
    valid_locations = [loc for loc in locations if pd.notna(loc)]

    k = random.randint(1, max_locations)          # Choose 1 to 3
    selected = random.sample(valid_locations, k)
    return ", ".join(selected)


# 3. Generate Data Frame
mdata = {
    "job_title": [random.choice(job_titles) for _ in range(num_rows)],
    "salary": [random.choice(salaries) for _ in range(num_rows)],


    "location": [get_random_locations() for _ in range(num_rows)],
    "compnay_size": [random.choice(sizes) for _ in range(num_rows)],
    "expereince_level": [random.choice(experience) for _ in range(num_rows)],
    "employment_type": [random.choice(employment) for _ in range(num_rows)],
    "skills": [get_messy_skills() for _ in range(num_rows)]
}

df = pd.DataFrame(mdata)

# 4. Inject Chaos (Shuffling values between columns)
for i in range(num_rows):
    if random.random() < 0.01: # 1% chance to shuffle row values
        cols = list(df.columns)
        shuffled = [df.at[i, c] for c in cols]
        random.shuffle(shuffled)
        for idx, c in enumerate(cols):
            df.at[i, c] = shuffled[idx]

# Save to CSV
df.to_csv("messy_data_analyst_jobs.csv", index=False)
mdata = pd.read_csv("/content/messy_data_analyst_jobs.csv")
print("Dataset generated: messy_data_analyst_jobs.csv")
```

## Data Cleaning

#### Clean the Skills Column

starting with the skills column
pretending that I don't know all values in the skills column
1. I need to find all values in the skills columns, then put them into a list
2. Start cleaning the the skills column,
   - start removing the special character such as `[` and `]`
   - remove all `NaN` and `Na` values and replace them with `None`

```python
# Initial skills list
skills_list = ['R', 'AWS', 'Azure', 'SQL', 'Python', 'Excel', 'BI',
               'Google Cloud', 'Tableau']

# Use a set for fast duplicate checking
skills_set = set(skills_list)

# Iterate through the column and add new skills
for value in mdata['skills']:
    if pd.isna(value):           # Skip empty/NaN values
        continue
    if value in skills_list:
        skills_list.set.discard("[ ","]")
        continue
    # Only add if it doesn't exist
        skills_set.add(value)
        skills_list.append(value)   # Keep the ordered list if you want

print(skills_list)
print(f"\nTotal unique skills: {len(skills_list)}")  # We now have all the skills that in skills columns
```

```python
# Create a function to filiter out the special character
def clean_skills(text):
  if pd.isna(text):
    return None     # returns a None value
  text = str(text)
  text = text.replace('[', '').replace(']', '').strip()   # remove '[' and ']' removes white space
  text = text.replace("'", "").replace('"', '')            # remove ' ' and " " removes white space
  text = text.replace('|', ',').strip()                    # remove '|' and replace with  ,
  text = text.replace('//', ',')                           # remove '/' and replace with  ,
  text = text.replace('/', ',')                            # remove '/' and replace with  ,
  text = text.replace('and', ',').strip()                  # remove 'and' removes white space
  text = text.replace(',,', ',')                           # remove any extra ',,' and replace with ','
  text = text.replace(';;', ',')                           # remove any extra ',,' and replace with ','
  text = text.replace(';', ',')                            # remove any extra ',,' and replace with ','
  return text

# Apply the function to the column
# mdata['skills'] = mdata['skills'].apply(clean_skills)
bdata['skills'] = bdata['skills'].apply(clean_skills)
```

```python
def clean_salary(text):
    if pd.isna(text):
        return 0

    text = str(text).strip()

    # Special cases for Words
    if any(x in text.upper() for x in ['DOE', 'Negotiable', 'Competitive']):
        return 0


    replacements = [
        # ('NaN', '0'),
        ('65000Competitive', '65000'),
        ('€', ''),
        ('$', ''),
        ('%', ''),
        ('120K', '120000'),
        ('120k', '120000'),
        ('70000 USD', '70000'),
        ('$50/hr', '104000'),     # 50 * 2080
        (',', ''),
        ('K', '000'),
        ('k', '000'),
    ]
    for old, new in replacements:
        text = text.replace(old, new)

    # Keep only digits
    text = ''.join(filter(str.isdigit, text))

    try:
        return int(text) if text else 0
    except:
        return 0

# Apply the function to the column
bdata['salary'] = mdata['salary'].apply(clean_salary)
```

```python
# Special Case : for two number stuck together
double_values = ['85000105000', '6000080000']

# Create a function that targets column
def concatenated_salary(salary):
    salary_str = str(salary).strip()

    if salary_str == '85000105000':
        num1 = 85000
        num2 = 105000
        val = statistics.median([num1, num2])
        return val

    elif salary_str == '6000080000':
        num1 = 60000
        num2 = 80000
        val = statistics.median([num1, num2])
        return val
    else:
        return salary_str

# Apply the fix
bdata['salary'] = bdata['salary'].apply(concatenated_salary)
```

### Cleaning the Company Size

To clean the company size column, we


1.  Rename `compnay_size` to `company_size`
2.  Come up with a way to distinguish the length of the company size
3.  Remove any special character from column such as `-`,`+`,`
4.  Remove any words such as `Startup`,`Unknown`,`Medium` and `Large`



```python
bdata['company_size'].head(100)
```

```python
# Create a function to clean the Company Column
def clean_company_size(company_size):

    # Map any words such as 'Startup', 'Unknown','Medium' and 'Large'
    mapping = {
        '501-1000': 'Large',
        '500': 'Small',
        '1-10': 'Startup',
        '10000+': 'Huge'
    }
    # Return the mapped value or the original if not in mapping
    return mapping.get(str(company_size), company_size)

# Apply the function to bdata dataframe
bdata['company_size'] = bdata['company_size'].apply(clean_company_size)
```

```python
# Create a new column called 'company_size_value' based on `company_size'
# Map new values
mapping = {'Unknown': 0, 'Startup': 1, 'Medium': 2, 'Large': 3, 'Huge': 4}

# # Create a new column in bdata and map values to 'expereince_level_values`
bdata['company_size_value'] = bdata['company_size'].map(mapping).fillna(0).astype(int)
```

### Cleaning Employment Type

To clean the `employment_type` column

1. Generate a list of `employment_type` to convert to number values
2. Convert `FT` to `Full Time`
3. Convert `C2C` to
4. Convert `PT` to `Part-time`
5. Convert `W2` to `Reguler`

```python
# Create a function for employment_type
def clean_employment_type(employment_type):
    # Mapping from abbreviations to full names
    mapping = {
        'FT': 'Full-time',
        'PT': 'Part-time',
        'W2': 'Regular',
        'C2C': 'Contract'
    }

    # First apply mapping (if it exists)
    cleaned = mapping.get(employment_type, employment_type)

    # Allowed values after cleaning
    allowed_types = ['Full-time', 'Part-time', 'Regular', 'Contract', 'Freelance', 'Temporary']


    # If value (after mapping) is allowed, keep it. Otherwise return None.
    if cleaned in allowed_types:
        return cleaned
    else:
        return None

# Apply the function
bdata['employment_type'] = bdata['employment_type'].apply(clean_employment_type)
```
