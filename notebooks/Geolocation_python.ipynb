```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import time
```

```python
# Your corrected list
cities = [
    'Chicago, IL', 'Indianapolis, IN', 'Philadelphia, PA', 'Denver, CO',
    'Nashville, TN', 'Las Vegas, NV', 'Louisville, KY', 'Norfolk, VA', # Fixed
    'Fargo, ND', 'Albuquerque, NM', # Fixed
    'New York, NY', 'San Francisco, CA', 'Dallas, TX', 'Austin, TX',
    'Houston, TX', 'Huntsville, AL', 'Fort Collins, CO', 'Salt Lake City, UT',
    'Phoenix, AZ', 'Orlando, FL', 'Santa Fe, NM', # Fixed
    'Jackson, MS', 'Detroit, MI', 'Atlanta, GA', 'Oklahoma City, OK', # Fixed
    'New Orleans, LA', 'Columbus, OH'
]

geod = pd.DataFrame({'city': cities})

# Initialize geolocator
geolocator = Nominatim(user_agent="my_geo_app")

# Define geocoding function with error handling
def get_location(address):
    try:
        location = geolocator.geocode(address)
        if location:
            return pd.Series([location.latitude, location.longitude])
        else:
            return pd.Series([None, None])
    except Exception as e:
        print(f"Error geocoding {address}: {e}")
        return pd.Series([None, None])

# Apply with rate limiting to avoid API bans (1 request per second)
geocode = RateLimiter(get_location, min_delay_seconds=1)
geod[['latitude', 'longitude']] = geod['city'].apply(geocode)

print(geod)
```
