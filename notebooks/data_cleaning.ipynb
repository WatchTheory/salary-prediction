````python
# Before Cleaning the Skills Column
bdata['skills'].head(10)
````

````python
# Initial skills list
skills_list = ['R', 'AWS', 'Azure', 'SQL', 'Python', 'Excel', 'BI',
               'Google Cloud', 'Tableau']

# Use a set for fast duplicate checking
skills_set = set(skills_list)

# Iterate through the column and add new skills
for value in mdata['skills']:
    if pd.isna(value):           # Skip empty/NaN values
        continue
    if value in skills_list:
        skills_list.set.discard("[ ","]")
        continue
    # Only add if it doesn't exist
        skills_set.add(value)
        skills_list.append(value)   # Keep the ordered list if you want

print(skills_list)
print(f"\nTotal unique skills: {len(skills_list)}")  # We now have all the skills that in skills columns
````

````python
# Create a function to filiter out the special character
def clean_skills(text):
  if pd.isna(text):
    return None     # returns a None value
  text = str(text)
  text = text.replace('[', '').replace(']', '').strip()   # remove '[' and ']' removes white space
  text = text.replace("'", "").replace('"', '')            # remove ' ' and " " removes white space
  text = text.replace('|', ',').strip()                    # remove '|' and replace with  ,
  text = text.replace('//', ',')                           # remove '/' and replace with  ,
  text = text.replace('/', ',')                            # remove '/' and replace with  ,
  text = text.replace('and', ',').strip()                  # remove 'and' removes white space
  text = text.replace(',,', ',')                           # remove any extra ',,' and replace with ','
  text = text.replace(';;', ',')                           # remove any extra ',,' and replace with ','
  text = text.replace(';', ',')                            # remove any extra ',,' and replace with ','
  return text

# Apply the function to the column
# mdata['skills'] = mdata['skills'].apply(clean_skills)
bdata['skills'] = bdata['skills'].apply(clean_skills)
````

````python
"""
Cleans the Salary column by remove an special characte

"""
def clean_salary(text):
    if pd.isna(text):
        return 0

    text = str(text).strip()

    # Special cases for Words
    if any(x in text.upper() for x in ['DOE', 'Negotiable', 'Competitive']):
        return 0

    # Remove special characters from the salary column
    replacements = [
        ('NaN', '0'),
        ('65000Competitive', '65000'),  # Remove '65000Competitive' and replace with '65000'
        ('€', ''),
        ('$', ''),  # Removes the '$' from the column and delete extra space
        ('%', ''),    # and so...
        ('120K', '120000'),
        ('120k', '120000'),
        ('70000 USD', '70000'),
        ('$50/hr', '104000'),     # 50 * 2080
        (',', ''),
        ('K', '000'),
        ('k', '000'),
    ]
    for old, new in replacements:
        text = text.replace(old, new)

    # Keep only digits
    text = ''.join(filter(str.isdigit, text))

    try:
        return int(text) if text else 0
    except:
        return 0

# Apply the function to the column
bdata['salary'] = mdata['salary'].apply(clean_salary)
````

````python
# Special Case : for two number stuck together
double_values = ['85000105000', '6000080000']

# Create a function that targets column
def concatenated_salary(salary):
    salary_str = str(salary).strip()

    if salary_str == '85000105000':
        num1 = 85000
        num2 = 105000
        val = statistics.median([num1, num2])
        return val

    elif salary_str == '6000080000':
        num1 = 60000
        num2 = 80000
        val = statistics.median([num1, num2])
        return val
    else:
        return salary_str

# Apply the fix
bdata['salary'] = bdata['salary'].apply(concatenated_salary)
````

````python
# Create a function to clean the Company Column
def clean_company_size(company_size):

    # Map any words such as 'Startup', 'Unknown','Medium' and 'Large'
    mapping = {
        '501-1000': 'Large',
        '500': 'Small',
        '1-10': 'Startup',
        '10000+': 'Huge'
    }
    # Return the mapped value or the original if not in mapping
    return mapping.get(str(company_size), company_size)

# Apply the function to bdata dataframe
bdata['company_size'] = bdata['company_size'].apply(clean_company_size)
````

````python
# Create a function for employment_type
def clean_employment_type(employment_type):
    # Mapping from abbreviations to full names
    mapping = {
        'FT': 'Full-time',
        'PT': 'Part-time',
        'W2': 'Regular',
        'C2C': 'Contract'
    }

    # First apply mapping (if it exists)
    cleaned = mapping.get(employment_type, employment_type)

    # Allowed values after cleaning
    allowed_types = ['Full-time', 'Part-time', 'Regular', 'Contract', 'Freelance', 'Temporary']


    # If value (after mapping) is allowed, keep it. Otherwise return None.
    if cleaned in allowed_types:
        return cleaned
    else:
        return None

# Apply the function
bdata['employment_type'] = bdata['employment_type'].apply(clean_employment_type)
````
